```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.ensemble import RandomForestRegressor
```

```python
df_paths = [
    '/gdrive/MyDrive/MACHINE LEARNING/ICO Coffee Dataset (Worldwide)/domestic-consumption.csv',
    '/gdrive/MyDrive/MACHINE LEARNING/ICO Coffee Dataset (Worldwide)/exports-calendar-year.csv',
    '/gdrive/MyDrive/MACHINE LEARNING/ICO Coffee Dataset (Worldwide)/exports-crop-year.csv',
    '/gdrive/MyDrive/MACHINE LEARNING/ICO Coffee Dataset (Worldwide)/gross-opening-stocks.csv',
    '/gdrive/MyDrive/MACHINE LEARNING/ICO Coffee Dataset (Worldwide)/total-production.csv'
]
```

```python
dfs = [pd.read_csv(df_path) for df_path in df_paths]
```

```python
def get_means(df):
    df = df.copy()
    countries = df[df.columns[0]]
    means = df.mean(axis=1)
    df = pd.concat([countries, means], axis=1)
    df.columns = ['country', countries.name]
    return df
```

```python
def make_df(dfs):
    
    # Process all DataFrames
    processed_dfs = []
    
    for df in dfs:
        processed_dfs.append(get_means(df))
        
    # Merge DataFrames
    df = processed_dfs[0]
    
    for i in range(1, len(processed_dfs)):
        df = df.merge(processed_dfs[i], on='country')
    
    return df
```

```python
data = make_df(dfs)
data
```

```python
def preprocess_inputs(df):
    df = df.copy()
    
    # Drop country column
    df = df.drop('country', axis=1)
    
    # Split df into X and y
    y = df['total_production']
    X = df.drop('total_production', axis=1)
    
    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, shuffle=True, random_state=1)
    
    # Scale X
    scaler = StandardScaler()
    scaler.fit(X_train)
    X_train = pd.DataFrame(scaler.transform(X_train), index=X_train.index, columns=X_train.columns)
    X_test = pd.DataFrame(scaler.transform(X_test), index=X_test.index, columns=X_test.columns)
    
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = preprocess_inputs(data)
```

```python
X_train
```

```python
y_train
```

```python
model = RandomForestRegressor()
model.fit(X_train, y_train)
print("Model trained.")
```

```python
y_pred = model.predict(X_test)

rmse = np.sqrt(np.mean((y_test - y_pred)**2))
print("RMSE: {:.2f}".format(rmse))

r2 = 1 - (np.sum((y_test - y_pred)**2) / np.sum((y_test - y_test.mean())**2))
print("R^2: {:.5f}".format(r2))
```
